```python
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import functions


#make folder for figs and data
def make_folder(names):
    '''makes folder for results and figs.
    names = list of folder names
    '''
    for _ in names:
        try:
            os.makedirs(os.path.join(os.getcwd(),_,''))
        except FileExistsError:
            pass

make_folder(['results/Accs_length/new'])

```

```python
import opt
rand_states = np.random.RandomState(12345)
plfold_args='-W 210 -u 50 -O'
pET21_UTR = 'GGGGAATTGTGAGCGGATAACAATTCCCCTCTAGAAATAATTTTGTTTAACTTTAAGAAGGAGATATACAT'


def df_primer_opt(df):
    df['Optimised_seq'] = df['optimiser'].apply(lambda x:x.simulated_anneal(rand_state=rand_states))
    return df
```

```python
#5 codons

failed_all = pd.read_csv('results/failed_optimised_SA.csv')
failed = failed_all[['Accession', 'Sequence']].copy()
#failed.head()

failed['optimiser'] = failed['Sequence'].apply(lambda x:opt.Optimizer(seq=x, utr=pET21_UTR, plfold_args=plfold_args,\
                                                                     ncodons=5))
failed_optimised5 = functions.parallelize_df(failed, df_primer_opt)
failed_optimised5.to_csv('results/Accs_length/new/optimised5.csv', index=None)



```

### Kolmogorov-Smirnov Test

```python
from scipy.stats import ks_2samp
```

```python
u = success['Accessibility'] #success
v = failed['Accessibility'] #failed
w = failed_optimised5_['Accessibility5'] #5
x = failed_optimised10_['Accs_opt_sa'] #10
y = failed_optimised15_['Accessibility15'] #15
z = failed_optimisedall_['Accessibilityall'] #all


labels = ['failed','success','four', 'nine','fourteen','full' ]
grps =  [v,u,w,x,y,z]
samples = pd.DataFrame(columns=labels)

df_ks = pd.DataFrame()
df_p = pd.DataFrame()
for ind,x in enumerate(samples.columns):
    for ind1,y in enumerate(samples.columns):
        #print(ind, ind1)
        ks1 = ks_2samp(grps[ind],grps[ind1])
        #print(ks1)
        df_ks.loc[x,y] = ks1[0]
        df_p.loc[x,y] = ks1[1]
df_ks.round(4)
```

```python
u.shape[0] + v.shape[0], u.shape, v.shape
```

```python

hmap_ks = df_ks[['success','failed','four', 'nine','fourteen','full' ]].copy()



mask = np.zeros_like(df_ks, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(3.75,2.75))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 240, s=90, l=25, as_cmap=True)
#cmap = sns.cubehelix_palette(8, start=.5, rot=-.75, as_cmap=True)
#cmap = sns.cubehelix_palette(50, as_cmap=True)


# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(df_ks, mask=mask, cmap=cmap, vmax=.75, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt=".2f")
ax.set_xlim(0,5)
ax.set_ylim(6,1)
plt.savefig("figs/kolmogorov_smirnov_distance_.pdf", bbox_inches = 'tight', pad_inches=0)
```

```python
cirt_df = pd.DataFrame() #critical values of KS statistic.

for ind,x in enumerate(samples.columns):
    for ind1,y in enumerate(samples.columns):
        crit_s = 1.63*np.sqrt((len(grps[ind])+len(grps[ind1]))/(len(grps[ind])*len(grps[ind1])))
        cirt_df.loc[x,y] = crit_s
cirt_df.round(4)
```
