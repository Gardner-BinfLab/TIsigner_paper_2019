```python
#make sure you install xlrd package first. This can be done by 'pip3 install --user xlrd'

import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



#make folder for figs and data
def make_folder(names):
    '''makes folder for results and figs.
    names = list of folder names
    '''
    for _ in names:
        try:
            os.makedirs(os.path.join(os.getcwd(),_,''))
        except FileExistsError:
            pass

make_folder(['figs/supp'])

```

```python
xls = pd.ExcelFile('1-s2.0-S1046202315301079-mmc2.xlsx')
codon_dict = xls.parse(2) #third sheit of xls has dicodon frequencies
codon_dict.head()
```

```python
all_dicodons = codon_dict['Unnamed: 2'][2:].values
genome = codon_dict['Unnamed: 3'][2:].values
transcr_top10 = codon_dict['Unnamed: 4'][2:].values
transcr_low10 = codon_dict['Unnamed: 5'][2:].values
transl_top10 = codon_dict['Unnamed: 6'][2:].values
prot_top10 = codon_dict['Unnamed: 7'][2:].values
prot_low10 =  codon_dict['Unnamed: 8'][2:].values
```

```python
dicod_dict = pd.DataFrame({'Dicodons':all_dicodons, 'Genome':genome, 'transcr_top10':transcr_top10,\
                          'transcr_low10':transcr_low10,'transl_top10':transl_top10,\
                          'prot_top10':prot_top10, 'prot_low10':prot_low10})
dicod_dict.head()
```

```python
dicod_dict.to_csv('dicodon_dic.csv', index=None)
```

```python

from collections import Counter
import numpy as np


dicod_dict =  pd.read_csv('dicodon_dic.csv', index=None)
def dicod(seq):
    assert len(seq)%3 == 0
    split_func = lambda seq, n: [seq[i:i+n] for\
                                    i in range(0, len(seq) - len(seq)%3 , n)]
    codons =  split_func(seq, 3)
    dicodons = []
    for index, val in enumerate(codons):
        try:
            dicodons.append(val+codons[index+1])
        except IndexError :
            pass
    
    #cnt_dict =Counter(dicodons)
    #summ=0
    #dic = []
    #for k, v in enumerate(cnt_dict.items()):
    #    summ+=v[1]
    #    dic.append(v[0])
    return dicodons


def freq_dic(seq):
    '''Codon adaptation index
    '''
    
    dicodon_list = dicod(seq)
    #print(dicodon_list)
    scores = []
    for _, val in enumerate(dicod_dict.columns[1:]):
        dicodon_freq = [np.log(dicod_dict.loc[dicod_dict['Dicodons'] == dicodon][val].values[0]) for dicodon 
                      in dicodon_list if dicod_dict.loc[dicod_dict['Dicodons'] == dicodon][val].values != 0]
        scores.append(np.exp(np.mean(dicodon_freq)))
    
    return scores
```

```python
data = pd.read_csv('results/features_and_merged_with_ixnos.csv')
```

```python
data.head()
```

```python
tt_seq = data[['Accession', 'Sequence', 'Class', 'Label']]
```

```python
tt_seq['dicodon_frq'] = tt_seq['Sequence'].apply(freq_dic)
```

```python
tt_seq.head()
```

```python
tt_seq[['Genome','transcr_top10','transcr_low10','transl_top10','prot_top10'\
     ,'prot_low10']] = pd.DataFrame(tt_seq.dicodon_frq.values.tolist(), index= tt_seq.index)
#tt_seq.to_csv('tt_seq_with_dicodon_freqs.csv', index=None)
tt_seq.head()
```

```python
tt_seq.to_csv('tt_seq_with_dicodon_freqs.csv', index=None)
```

```python
import seaborn as sns
from sklearn.metrics import roc_curve, auc

import matplotlib.pyplot as plt



df = tt_seq[['Genome',
'transcr_top10',
'transcr_low10',
'transl_top10',
'prot_top10',
'prot_low10' ]].copy()

labels = tt_seq.Label.values
plt.figure(0).clf()
lw = 1.5

fig = plt.figure(figsize=(4, 8))

sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.4)
sns.set_palette("bright")
plt.subplot(2, 1, 2)
for i, col in enumerate(df.columns):
    preds = df[col].values
    fpr, tpr, _ = roc_curve(labels, preds)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, # c=np.random.rand(3,),
             lw=lw, label=col +' (%0.4f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.title('Codon context from various sources')
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        labelbottom=False) # labels along the bottom edge are off

# Make common axis labels
fig.text(0.5, 0.04, 'False positive rate (1-Specificity)', va='center', ha='center')
fig.text(-0.05, 0.30, 'True positive rate (Sensitivity)', va='center', ha='center', rotation='vertical')

sns.despine()
plt.savefig("figs/supp/ROC_codon_pair.pdf", bbox_inches = 'tight', pad_inches = 0)
plt.show()
```
